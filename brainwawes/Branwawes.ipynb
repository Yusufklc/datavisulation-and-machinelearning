```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sys
from sklearn.preprocessing import LabelBinarizer,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score ,accuracy_score
```

```python

data = pd.read_csv("DataSets/acquiredDataset.csv")


def relplots(colname):   
    plt.subplot(1,2,1)
    plt.scatter(x = "classification",y=colname,data=data[data["classification"]==0])
    plt.scatter(x = "classification",y=colname,data=data[data["classification"]==1])
    plt.subplot(1,2,2)
    sys.boxplot(x="classification",y=colname,data=data)
    plt.show()
    
relplots("attention")
```

```python

for i in ["highGamma","highBeta","highAlpha","lowAlpha","theta","delta"]:
    data[i] = StandardScaler().fit_transform(data[[i]])

data["classification"] = LabelBinarizer().fit_transform(data["classification"])
    
X = data[["highGamma","highBeta","highAlpha","lowAlpha","theta","delta"]]

y = data["classification"]

Xtrain , Xtest ,Ytrain ,Ytest = train_test_split(X,y,test_size=0.2)

model = KNeighborsClassifier()

model.fit(Xtrain,Ytrain)

prediction = model.predict(Xtest)

acc = accuracy_score(prediction,Ytest)

acc2 = f1_score(prediction,Ytest)

acc,acc2
```
